```python
import numpy as np
import matplotlib.pyplot as plt

def plot_two_frequencies(f1=10, f2=100):
    # Definir el rango de t
    t = np.linspace(0, 0.5, 1000)
   
    # Calcular las funciones seno
    y = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)
   
    # Configurar el estilo de la grfica
    plt.style.use('ggplot')
   
    # Crear la grfica
    plt.figure(figsize=(10, 6))
    plt.plot(t, y, label=f'sen(2{f1}t) + sen(2{f2}t)', color='b')
   
    # Aadir leyenda
    plt.legend()
   
    # Aadir ttulo y etiquetas
    plt.title('Dos frecuencias')
    plt.xlabel('Tiempo (t)')
    plt.ylabel('Amplitud')
   
    # Mostrar la grfica
    plt.show()

# Llamar a la funcin con los valores predeterminados
plot_two_frequencies()

# Llamar a la funcin con valores personalizados
# plot_two_frequencies(f1=20, f2=200)
```
