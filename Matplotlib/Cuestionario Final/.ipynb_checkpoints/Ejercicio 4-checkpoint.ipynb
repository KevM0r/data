```python
import matplotlib.pyplot as plt
import numpy as np

# Crear un array de valores 100 valores que van de 0 a 2pi
x = np.linspace(0, 2 * np.pi, 100)

# Calcular las funciones seno y coseno
yseno = np.sin(x)
ycoseno = np.cos(x)

# Crear una figura y un conjunto de subplots con 2 filas y 1 columna
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Graficar la función seno en el primer subplot
ax1.plot(x, yseno)
ax1.set_title('Función seno')

# Graficar la función coseno en el segundo subplot
ax2.plot(x, ycoseno)
ax2.set_title('Función coseno')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
