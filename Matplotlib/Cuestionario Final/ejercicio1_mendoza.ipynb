##Ejercicio 1
La tarea más habitual a la hora de trabajar con matplotlib es representar una función. Lo que tendremos que hacer es definir un dominio y evaluarla en dicho dominio. Por ejemplo:
$$ f(x) = e^{-x^2} $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return np.exp(-x**2)

# Crear un array de valores x con 400 puntos para mayor presicion
x = np.linspace(-3, 3, 400)

# Calcular los valores y
y = f(x)

# Crear el gráfico
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="$f(x)$ = $e^{-x^2}$")
plt.title("Gráfico de $f(x)$ = $e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
