Con parametros <span style='background-color: #F5F5F5F5; padding:3px 6px;'>s</span> y <span style='background-color: #F5F5F5F5; padding:3px 6px;'>c</span> podemos modificar el tamaño y el color respectivamente. Para el color, a cada valor numérico se le asigna un color a través de un <strong>mapa de colores</strong>; ese mapa se puede cambiar con el argumento <span style='background-color: #F5F5F5F5; padding:3px 6px;'>cmap</span>. Esa correspondencia se puede visualizar llamando a la función <span style='background-color: #F5F5F5F5; padding:3px 6px;'>colorbar</span>.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 50 numeros aleatorios del 0 al 900

# Crear el gráfico de dispersión
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # mostrar la barra de colores
plt.show()
```

<h4>Usando mapa de colores 'plasma' y valor de transparencia 0.8</h4>

```python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)  
area = (30 * np.random.rand(N))**2  # 50 numeros aleatorios del 0 al 900

# Crear el gráfico de dispersión
plt.scatter(x, y, s=area, c=colors, alpha=0.8, cmap='plasma')
plt.colorbar()  # mostrar la barra de colores
plt.show()
```
