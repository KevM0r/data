Investigar :

$1 - f(x)$

https://matplotlib.org/2.0.2/api/pyplot_api.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x)
def f(x):
    return np.exp(-x**2)

# Crear un rango de valores de x
x = np.linspace(-1,3,400)

# Calcular los valores de f(x) y 1-f(x)
y1 = f(x)
y2 = 1 - f(x)

# Crear gráfica
plt.figure()

# Graficar funciones f(x) y 1 - f(x)
plt.plot(x, y1, 'ro', label='$f(x)=e**{-x**2}$')  # Gráfica con puntos rojos
plt.plot(x, y2, 'g--', label='$1 - f(x)$')        # Gráfica con línea verde discontinua

# Personalizar la gráfica
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()  # Crear una leyenda

# Mostrar gráfica
plt.show()
```
